```python
import pandas as pd
```

```python
data = pd.read_csv('World Weather Repository.csv')
```

```python
data
```

```python
data = data.dropna() # Removing the rows which contain null values
```

```python
data.info()
```

```python
data.describe()
```

```python
X = data[['humidity', 'wind_mph', 'pressure_mb', 'precip_mm', 'cloud', 'feels_like_celsius', 'visibility_km', 'uv_index']] # This are the features Columns
```

```python
X.head()
```

```python
X.describe()
```

```python
y = data['temperature_celsius'] # This is the target column
```

```python
y.head()
```

```python
y.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plots for a few features
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
sns.scatterplot(x='humidity', y='temperature_celsius', data=data)
plt.title('Humidity vs Temperature')

plt.subplot(2, 2, 2)
sns.scatterplot(x='wind_mph', y='temperature_celsius', data=data)
plt.title('Wind Speed vs Temperature')

plt.subplot(2, 2, 3)
sns.scatterplot(x='pressure_mb', y='temperature_celsius', data=data)
plt.title('Pressure vs Temperature')

plt.subplot(2, 2, 4)
sns.scatterplot(x='feels_like_celsius', y='temperature_celsius', data=data)
plt.title('Feels Like vs Temperature')


plt.tight_layout()
plt.show()
```

```python
correlation_matrix = pd.concat([X, y], axis=1).corr()

# Creating a heatmap to see correaltion between X and Y
plt.figure(figsize=(10, 8))  # Adjust figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

Evluation

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print(f"R-squared: {r2}")
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
```
